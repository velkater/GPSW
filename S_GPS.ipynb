# Operation $S$ and GPS words

```python
import gpc
import itertools
```

```python
# Number of steps
steps = 4
# Combinations of delta and theta
thetas = [''.join(i) for i in itertools.product('ER', repeat=steps)]
deltas = [''.join(i) for i in itertools.product('01', repeat=steps)]
```

```python
for delta in deltas:
    for theta in thetas:
        u = gpc.makeWord(delta, theta, steps, seed = "001000100")
        resu = gpc.isGPS(u)
        res = gpc.isGPS(gpc.makeS(u))
        if res[0] == True:
            print("U:", u[:40],"...,", "pridani:", delta, theta)
            print("S(U):", gpc.makeS(u)[:40], "...",)
            print("Norm. posl. U:", [resu[1], resu[2]])
            print("Norm. posl. S(U):", [res[1], res[2]])    
            print("Max. R posl S(U):", gpc.maximizeRinBiseq(res[1],res[2]))
            print("")
```
