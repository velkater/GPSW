```python
import itertools
import zpu as zpu
```

```python
def subs(dic, n):
    s = dic["0"]
    i = 1
    while len(s) < n:
        s = s + dic[s[i]]
        i = i + 1
    return s[:n]
```

```python
dictio = {"0": "01", "1":"0"}
t = subs(dictio, 20)
t
```

```python
phi0 = ['0'+ ''.join(i) for i in itertools.product('01', repeat=1)]
phi1 = [''.join(j) for j in itertools.product('01', repeat=4)]
cart = [ k for k in itertools.product(phi0, phi1)]
print(phi0)
print(phi1)
```

```python
n = 100
for rule in cart:
    dicti = {"0": rule[0], "1": rule[1]}
    word = subs(dicti, n)
    ret = zpu.isZps(word)
    print(word)
    print(dicti, ret)
```
